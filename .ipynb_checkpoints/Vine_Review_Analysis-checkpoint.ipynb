```python
import pandas as pd
```

```python
#importing dataset from csv
vine_table_df = pd.read_csv("vine_table.csv")
vine_table_df.head()
```

```python
#filtering dataset for total_votes greater or equal to 20; putting into new dataframe
above_twenty_df = vine_table_df[vine_table_df["total_votes"] >= 20]
above_twenty_df.head()
```

```python
#creating new dataframe and adding a "percent_helpful" column
helpful_votes_df = above_twenty_df
helpful_votes_df["percent_helpful"] = helpful_votes_df["helpful_votes"] / helpful_votes_df["total_votes"]
helpful_votes_df
```

```python
#filtering the percent_helpful column to be greater or equal to 50%
helpful_votes_df = helpful_votes_df[helpful_votes_df["percent_helpful"] >= 0.5]
helpful_votes_df
```

```python
#filtering for reviews that are part of the vine program; putting these into a new dataframe
vine_reviews_df = helpful_votes_df[helpful_votes_df["vine"] == 'Y']
vine_reviews_df
```

```python
#filtering for reviews that are not part of the vine program; putting these into a new dataframe
non_vine_reviews_df = helpful_votes_df[helpful_votes_df["vine"] == 'N']
non_vine_reviews_df
```

```python
vine_review_count = vine_reviews_df["vine"].count()
vine_fivestar_review_count = vine_reviews_df[vine_reviews_df["star_rating"] == 5]["vine"].count()
vine_fivestar_percentage = vine_fivestar_review_count / vine_review_count
print(f"There are {vine_review_count} Vine program reviews, {vine_fivestar_review_count} of which are five star reviews. Five star reviews make up {vine_fivestar_percentage*100:.1f}% of the Vine reviews.")
```

```python
non_vine_review_count = non_vine_reviews_df["vine"].count()
non_vine_fivestar_review_count = non_vine_reviews_df[non_vine_reviews_df["star_rating"] == 5]["vine"].count()
non_vine_fivestar_percentage = non_vine_fivestar_review_count / non_vine_review_count
print(f"There are {non_vine_review_count} unpaid reviews, {non_vine_fivestar_review_count} of which are five star reviews. Five star reviews make up {non_vine_fivestar_percentage*100:.1f}% of the unpaid reviews.")
```
